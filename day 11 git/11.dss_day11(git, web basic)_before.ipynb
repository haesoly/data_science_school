git 기본 명령어 

 - git init
  + 새로운 깃 저장소 추가(깃으로 관리할 수 있는 폴더가 됨)
  
 - git add [file]
  + 스테이지(준비영역)에 변경 이력 커밋 준비
  
 - git commit -m "message"
  + HEAD(로컬저장소)에 변경이력 추가
  
 - git push origin [branch]
  + 원격 저장소에 추가
  
 - git pull (git fetch + git merge)
  + 원격 저장소에서 최신 이력을 받아와, 로컬저장소와 작업공간의 3 way merge
 
 - git fetch + git rebase
  + 원격 저장소에서 최신 이력을 받아와, 로컬저장소에 rebase 

* git branch

* commit 삭제하기

* bs4 (BeautifulSoup) module
 - 가장 널리 쓰이는 html parsing 모듈
 - html 각 요소의 값을 다양한 방법으로 검색가능

* Web page html 파헤치기
 - chrome, IE 등 브라우져의 개발자 도구를 이용하여 분석

```python
import requests
from bs4 import BeautifulSoup

def get_news_title(url):
    response = requests.get(url)
    content = response.text
    
    soup = BeautifulSoup(content)
    title = soup.title.get_text()  #title태그 접근
  
    return title


print get_news_title('http://media.daum.net/foreign/newsview?newsid=20160921114543616')


def get_news_content(url):
    response = requests.get(url)
    content = response.text

    soup = BeautifulSoup(content)
#soup.div -> 첫번째 div만 가져옴
#soup.find 사용
#soup.find_all('p')
    div = soup.find('div', attrs = {'id' : 'harmonyContainer'})
    #div = soup.find('div', attrs = {'class' : 'article_view'})
    content = ''
    for paragraph in div.find_all('p'):  #find_all은 list 반환 
        content += paragraph.get_text()
        
    return content.encode('utf-8')
        
news1 = get_news_content('http://v.media.daum.net/v/20170112152314639')
news2 = get_news_content('http://media.daum.net/digital/newsview?newsid=20160920180606199')    

print news1
print news2
```
